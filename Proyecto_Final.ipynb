```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/DGeanf/Calificacion_Chocolate/main/chocolate_ratings.csv'
datos = pd.read_csv(url, on_bad_lines='error')
datos.shape #Cuantos datos hay
```

```python
datos.head()
```

```python
# Revisando la descripcion
datos.describe(include='all')
```

```python
# Revisando los tipos de datos
datos.dtypes
```

```python
## Renombraremos las columnas para un mejor manejo
original_col_nombres = datos.columns
nuevo_col_nombres = ['REF', 'empresa', 'pais_empresa', 'año_analisis', 'cacao_pais', 'lugar_cacao',
                'cacao_porcentaje', 'ingredientes', 'caract_principal', 'rating']
datos = datos.rename(columns=dict(zip(original_col_nombres, nuevo_col_nombres)))
## Modificando el tipo de dato
datos['cacao_porcentaje'] = datos['cacao_porcentaje'].str.replace('%','').astype(float)/100
datos.head()
```
